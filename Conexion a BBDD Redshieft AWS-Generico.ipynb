```python
############ Requerimientos ####################
# sudo apt-get install python-pip 
# sudo apt-get install libpq-dev
# sudo pip install psycopg2
# sudo pip install sqlalchemy
# sudo pip install sqlalchemy-redshift
##############################################

# Importamos la biblioteca sqlalchemy
import sqlalchemy as sa
from sqlalchemy.orm import sessionmaker

#>>>>>>>> Introducir los datos de la conexion AQUI <<<<<<<<<<<<< 
DATABASE = ""
USER = ""
PASSWORD = ""
HOST = ""
PORT = "5439"
SCHEMA = "public"      #Por defecto es "public" 


####### Crear la sesion y conexion ############## 
connection_string = "redshift+psycopg2://%s:%s@%s:%s/%s" % (USER,PASSWORD,HOST,str(PORT),DATABASE)
engine = sa.create_engine(connection_string)
session = sessionmaker()
session.configure(bind=engine)
s = session()
SetPath = "SET search_path TO %s" % SCHEMA
s.execute(SetPath)
```

```python
#>>>>>>>> Escribir la consulta SQL AQUI <<<<<<<<<<<<< 

query = "SELECT * FROM ...;"
```

```python
# Ejecucion del cursor sin pd - NO NECESARIO PARA DATAFRAME
rr = s.execute(query)
all_results =  rr.fetchall()
```

```python
# Importamos la biblioteca para la creacion de dataframes
import pandas as pd
```

```python
# Creamos el DataFrame
df = pd.read_sql(query, engine)
```
